```python
def main_parties(df):
    votes = df[df.columns[11:]]
    sums = votes.sum() 
    sums[sums > 100000]
    return list(sums[sums > 100000].index)
```

```python
def remove_minor(df):
    return df[desired_cols + main_parties(df)]
```
